#**Import Library**

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("/content/mrbean.jpg")
img_real = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

# **Intensity Adjustment**

```python
brightness_factor = 1.5
adjusted_image = cv2.convertScaleAbs(image, alpha=brightness_factor, beta=1)
```

```python
plt.subplot(1, 2, 1)
plt.title("Original")
plt.imshow(img_real)
plt.subplot(1, 2, 2)
plt.title("Brightness & contrast")
plt.imshow(adjusted_image)
plt.show()
```

# **Histogram Equalization**

```python
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
equalized_image = cv2.equalizeHist(gray_image)

plt.subplot(1, 2, 1)
plt.title("Gray Image")
plt.imshow(gray_image,cmap='gray')

plt.subplot(1, 2, 2)
plt.title("Equalize")
plt.imshow(equalized_image,cmap='gray')
plt.show()
```

# **Threasholding**

```python
img = gray_image
assert img is not None, "file coould not be read, check with os.path.exist()"
ret,thresh1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img,127,255,cv2.THRESH_TOZERO_INV)
```

```python
titles = ['Original Image', 'Binary', 'THRESH_BINARY_INV', 'TRUNC', 'TOZERO', 'TOZERO_INV']
images = [img, thresh1, thresh2, thresh3, thresh4, thresh5]

for i in range(6):
  plt.subplot(2,3,i+1), plt.imshow(images[i], 'gray', vmin=0, vmax=255)
  plt.title(titles[i])
  plt.xticks([]), plt.yticks([])
plt.show()
```
